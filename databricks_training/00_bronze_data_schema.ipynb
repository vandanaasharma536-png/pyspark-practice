```python
catalog_name = "finance_md"
schema_name = "market_indices"
database1_name = "bronze"
# database2_name = "bse_it"
volumes = ["datasets", "checkpoints"]
```

```python
create_catalog = f"create catalog if not exists {catalog_name}"
# create_schema = f"create schema if not exists {schema_name}"
create_database1 = f"create database if not exists {database1_name}"
# create_database2 = f"create database if not exists {database2_name}"

```

```python

spark.sql(create_catalog)
spark.sql(f"use catalog {catalog_name}")

# spark.sql(create_schema)
# spark.sql(f"use schema {schema_name}")

assert(spark.sql("select current_catalog()").collect()[0][0] == catalog_name)
# assert(spark.sql("select current_schema()").collect()[0][0] == schema_name)
```

```python
# spark.sql(f"use schema {schema_name}")
spark.sql(create_database1)
# spark.sql(create_database2)
```
