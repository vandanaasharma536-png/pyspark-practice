```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DecimalType, DateType

delta_table_name = "finance_md.bronze.nifty_it_list"
delta_table_schema =  StructType([
  StructField("company_name", StringType(), False),
  StructField("symbol", StringType(), False)]
)
```

```python
df = spark.read.format("csv").schema(delta_table_schema).option("header", True).load(r"/Volumes/finance_md/bronze/datasets/nifty_it_list.csv")

df.display()
```

```python
df.write.format("delta").mode("overwrite").saveAsTable(delta_table_name)
```
